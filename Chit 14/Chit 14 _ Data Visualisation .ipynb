# HISTOGRAM 

```python
fig, axes = plt.subplots(2, 2, figsize=(16, 8))


axes[0,0].set_title("Distribution of Sepal Length")
axes[0,0].hist(df["sepal.length"]);

axes[0,1].set_title("Distribution of Sepal Width")
axes[0,1].hist(df["sepal.width"]);

axes[1,0].set_title("Distribution of Petal Length")
axes[1,0].hist(df["petal.length"]);

axes[1,1].set_title("Distribution of Petal Width")
axes[1,1].hist(df["petal.width"]);
```

# BOXPLOT

```python
sns.boxplot(x=df['sepal.length'])
#Median lies between 5.5 to 6
```

# COMPARISION

```python
fig, axes = plt.subplots(2, 2, figsize=(16,9))
axes[0,0].set_title("Distribution of Sepal Length")
sns.boxplot(  y="sepal.length", x= "variety", data=df,  orient='v' , ax=axes[0, 0])

axes[0,1].set_title("Distribution of Sepal Length")
sns.boxplot(  y="sepal.width", x= "variety", data=df,  orient='v' , ax=axes[0, 1])

axes[1,0].set_title("Distribution of Sepal Length")
sns.boxplot(  y="petal.length", x= "variety", data=df,  orient='v' , ax=axes[1, 0])

axes[1,1].set_title("Distribution of Sepal Length")
sns.boxplot(  y="petal.width", x= "variety", data=df,  orient='v' , ax=axes[1, 1])

plt.show()
```
